```python
train_data['Gender'] = train_data['Sex']
```

```python
train_data = pd.get_dummies(train_data, columns=['Gender'])
```

```python
train_data.head()
```

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Gender_female', 'Gender_male']
```

```python
test_data = pd.read_csv('test.csv')
```

```python
test_data.shape
```

```python
test_data.head()
```

```python
test_data['Age'] = test_data['Age'].fillna(value=age_average)
```

```python
test_data['Gender'] = test_data['Sex']
```

```python
test_data = pd.get_dummies(test_data, columns=['Gender'])
```

```python
test_data.head()
```

```python
trainX = train_data[features]
```

```python
trainY = train_data['Survived']
```

```python
testX = test_data[features]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=200, random_state=20)
```

```python
model.fit(trainX, trainY)
```

```python
prediction = model.predict(testX)
```

```python
prediction
```

```python
result = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': prediction})
```

```python
result.head()
```

```python
result.to_csv('submission.csv', index=False)
```
